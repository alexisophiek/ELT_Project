```python
import numpy as np
import csv
import json
import matplotlib.pyplot as plt; plt.rcdefaults()
import matplotlib.pyplot as plt
import matplotlib.pyplot as mlp
import pandas as pd
from pandas import DataFrame

from sqlalchemy import create_engine
```

```python
df = pd.read_csv('data.csv')
df = df.drop('URL', axis=1)
df['track'] = df['Artist']+ ' - ' + df['Track Name']
df.head()
```

```python
All_streams = df.groupby(['track'])['Streams'].sum().reset_index()
All = All_streams.sort_values(['Streams'], ascending=False)
All.rename(columns = {'Streams':'all_streams'}, inplace = True)

# create column "position"
All['all_position'] = range(1, 1+len(All))
export_csv = All.to_csv('all.streams.csv', index=False)
All.head(10)
```

```python
US_only_df = df[df['Region'] == 'us']
US_only_df.head()
```

```python
# groupby artist and track then stream count
US = US_only_df.groupby(['track'])['Streams'].sum().reset_index()
US_df = pd.DataFrame(US)
# sort descending order on stream count
US = US_df.sort_values(['Streams'], ascending=False)
US.rename(columns = {'Streams':'us_streams'}, inplace = True)
# create column "position"
US['us_position'] = range(1, 1+len(US))
# create index to value to find top 10, and delete the rest
indexNames = US[ (US['us_position'] >= 50) ].index
US.drop(indexNames , inplace=True)
# copy data
US.set_index('us_position')
# print(US.US_Streams.values.tolist())
# print(US.Position.values.tolist())
# print(US.Artist_and_Track_Name.values.tolist())

export_csv = US.to_csv('usa.streams.csv', index=False)
usa_streams = pd.read_csv('usa.streams.csv', index_col=2)
usa_streams.head(10)
```
