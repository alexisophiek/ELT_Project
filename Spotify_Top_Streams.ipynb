```python
# Dependencies
from sqlalchemy import create_engine
import pandas as pd
```

```python
df = pd.read_csv('Resources/data.csv')
df = df.drop('URL', axis=1)
# groupby artist and track then stream count
new_df = df.groupby(['Region','Artist','Track Name'])['Streams'].sum().reset_index()
# sort descending order on stream count
# new_df = df.sort_values(['Streams'], ascending=False)
new_df.rename(columns = {'Streams':'streams', 'Track Name':'track', 'Artist':'artist', 'Region':'region'}, inplace = True)
new_df
# # create column "Ranking"
# new_df['ranking'] = range(1, 1+len(new_df))
# # create index to value to find top 10, and delete the rest
spotify_streams = pd.DataFrame(new_df)
# spotify_streams.set_index(('ranking'), inplace=True)
spotify_streams
```

```python
connection_string = "postgres:postgres@localhost:5432/2017_StreamingDB"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
spotify_streams.to_sql(name='top_spotify', con=engine, if_exists='append', index=False)
```
