```python
# Dependencies
import requests 
from bs4 import BeautifulSoup as bs
from sqlalchemy import create_engine
import pandas as pd
```

```python
# Specify the URL
url = "https://www.billboard.com/charts/year-end/2017/streaming-songs"
# Make request and store response
response = requests.get(url)

# Print status code
response.text[:200]
```

```python
# Create a BeautifulSoup object
soup = bs(response.text, 'html.parser')
```

```python
# list_items = soup.find('article', class_="ye-chart-item")
year_end_list = soup.find_all('article', class_="ye-chart-item")
```

```python
def top_tracks_details(year_end_list):
    rank = year_end_list.find("div", class_="ye-chart-item__rank").text.strip()
    track = year_end_list.find("div", class_="ye-chart-item__title").text.strip()
    artist = year_end_list.find("div", class_="ye-chart-item__artist").text.strip()
    return {"rank":rank,
           "track":track,
           "artist":artist}
```

```python
top = [top_tracks_details(each) for each in year_end_list];
top
```

```python
bb_2017 = pd.DataFrame(top).set_index("rank")
bb_2017['artist_and_Track_Name'] = bb_2017['artist']+ ' - ' + bb_2017['track']

bb_2017.copy()
```
