```python
from textwrap import wrap
import psycopg2
import pandas as pd
import sqlalchemy
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rcParams
from sqlalchemy import create_engine
```

```python
# Necessary Postgres login information
connection_string = "postgres:postgres@localhost:5432/2017_StreamingDB"
# Create the connection
engine = create_engine(f'postgresql://{connection_string}')
```

```python
Top_BB = pd.read_sql_query('''SELECT *  FROM top_billboard WHERE rank < '11';''', engine)
Top_BB.set_index(('rank'), inplace=True)
#Rename Columns and Drop Columns
Top_BB['Track_by_Artist']= Top_BB['track'].values + ' - ' + Top_BB['artist']
Top_BB
```

```python
All_df = pd.read_sql_query('''SELECT artist, track,SUM (streams) FROM top_spotify GROUP BY artist,track;''', engine)
All_df = All_df.sort_values('sum', ascending=False)
All_df['spotify_rank'] = range(1, 1+len(All_df))
All_df.set_index(("spotify_rank"), inplace=True)
# All_df['Track_by_Artist']= All_df['track'].values + ' - ' + All_df['artist']
# All_df = All_df[['artist','track','spotify_rank','streams']]
All_df.head(10)
```
